###**Handling the 'Not Rated' rows for our Target Variable**

  Based on insights from Subha (2024) [Medium Article](https://medium.com/@pingsubhak/handling-missing-values-in-dataset-7-methods-that-you-need-to-know-5067d4e32b62), the best approach to handling these 'Not Rated' values—categorized as 'Missing at Random'—is to train a model using only the data with existing ratings and then predict ratings for the missing ones. The predicted values will then be used to fill in the missing ratings.



```python
#Creating a function to help achieve it. Note: I intend on using this function for my final prediction with a few tweaks.

#modules needed for Linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#modules needed for Random Forest.
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score


def Not_Rated_Predictor(df, features):
    """
    Predicts and fills zero-rated Aggregate ratings using the best-performing regression model.
    Compares Linear Regression and Random Forest using cross-validation and selects the best model.

    Parameters:
    a) df: The dataset containing restaurant information.
    b) features: List of features to train the model on.

    Returns the dataset with zero ratings imputed.
    """

    #Create a new column, not rated restaurant = 0, rated=1
    df['Has Rating'] = df['Aggregate rating'].apply(lambda x: 1 if x > 0 else 0)

    #Select relevant columns
    df_selected = df[features + ['Aggregate rating']].copy()   #.copy() ensures we avoid warnings in pandas related to working on a copy

    #Split data into Rated & Not Rated
    rated_rows = df_selected['Aggregate rating'] > 0
    not_rated_rows = df_selected['Aggregate rating'] == 0
    rated_df = df_selected[rated_rows]  # Training Data
    unrated_df = df_selected[not_rated_rows]  # Data to be filled

    #prevents unnecessary execution/crash when all restaurants are filled
    if unrated_df.empty:
        print("No missing ratings found. Returning original dataset.")
        return df

    #Train & Compare Models
    X_train = rated_df[features]
    X_unrated = unrated_df[features]
    y_train = rated_df['Aggregate rating']

    df_features = rated_df[features]

    #Seperate categorical and numerical columns
    num_features = df_features.select_dtypes(include=np.number).columns
    cat_features = df_features.select_dtypes(include='object').columns

    #Preprocessing
    from sklearn.preprocessing import OneHotEncoder, StandardScaler

    #encode all categorical data
    encoder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=True)

    X_train_cat = encoder.fit_transform(X_train[cat_features])
    X_unrated_cat = encoder.transform(X_unrated[cat_features])

    #scale numerical data
    scaler = StandardScaler()
    X_train_num = scaler.fit_transform(X_train[num_features])
    X_unrated_num = scaler.transform(X_unrated[num_features])

    ##Combine encoded features with scaled featured
    X_train_processed = np.hstack((X_train_cat.toarray(), X_train_num))
    X_unrated_processed = np.hstack((X_unrated_cat.toarray(), X_unrated_num))


    #Models
    models = {
        "Linear Regression": LinearRegression(),
        "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    }

    best_model = None
    best_score = float('inf')  # The Lower RME, the better

    for name, model in models.items():

        scores = cross_val_score(model, X_train_processed, y_train, scoring='neg_mean_squared_error', cv=5)
        # Convert MSE to RMSE
        rmse_score = np.sqrt(-scores.mean())

        print(f"{name} RMSE: {rmse_score:.4f}")

        #Select model with lowest RMSE
        if rmse_score < best_score:
            best_score = rmse_score
            best_model = model

    print(f"\nBest Model Selected: {best_model.__class__.__name__} with RMSE: {best_score:.4f}")

    #Train Best Model
    best_model.fit(X_train_processed, y_train)

    #Predict missing ratings
    predicted_ratings = best_model.predict(X_unrated_processed)

    #Fill missing ratings in the original dataset
    df.loc[df['Aggregate rating'] == 0, 'Aggregate rating'] = predicted_ratings

    print('"Not Rated" rows successfully filled using', best_model.__class__.__name__)
    df = df.drop(columns = 'Has Rating')
    print("\nReplacing Rating text & Rating Color for predicted ratings...............")

    #Define rating text and color mapping
    rating_mapping = {
      (0.0, 2.2): ("Poor", "Red"),
      (2.3, 3.3): ("Average", "Orange"),
      (3.4, 4.4): ("Very Good", "Green"),
      (4.5, 5.0): ("Excellent", "Dark Green")
      }

    #convert predicted ratings to series
    predicted_ratings = pd.Series(predicted_ratings)


    filt =  df['Aggregate rating'] == 0.0
    # Function to map rating text & color based on predicted rating
    def map_rating_info(rating):
      for (low, high), (text, color) in rating_mapping.items():
        if low <= rating <= high:
          return pd.Series([text, color])
      return pd.Series(["Not Rated", "White"])

    df.loc[filt, ['Rating text', 'Rating color']] = predicted_ratings.apply(map_rating_info)

    print("\nPrediction and Rating Update Complete.")

    return df
```

##### Function implementation

```python
#Define features to use for prediction excluding ('Aggregate rating')
features = [
    'Restaurant Name','City','Average Cost for two', 'Cuisines', 'Currency', 'Has Table booking',
    'Has Online delivery', 'Is delivering now', 'Switch to order menu','Rating text', 'Rating color',
    'Price range','Votes'
    ]
#function call to fill missing ratings
df_filled = Not_Rated_Predictor(df, features)

```

## **Geospatial Analysis**
Based on insights from [Lisa M. (2018) ](https://medium.com/@lisa.mitford/geo-spatial-analysis-with-python-fdddd69eebea) & [Haversine formula](https://stackoverflow.com/questions/27928/calculate-distance-between-two-latitude-longitude-points-haversine-formula), we will conduct Geospatial Analysis using the Longitude and Latitude columns. This analysis aims to extract meaningful location-based insights by:

1.   Creating a Geometrical DataFrame → Convert restaurant locations into a GeoDataFrame using GeoPandas.

2.   Generating Geometric Points → Represent each restaurant as a point (Longitude, Latitude).
3.  Evaluating Spatial Patterns →
*   DBSCAN Clustering: Identify restaurant hubs based on location density. (Because studies from Gboeing,2014 suggests Kmeans is not the best algorithm  for spatial data
*   Haversine Distance Calculation: Measure distances from restaurants to key locations, such as city centers or top-rated restaurants.

This geospatial analysis will help determine whether location influences restaurant ratings and identify key restaurant clusters.

```python
#import necessary libraries
import geopandas as gpd
from shapely.geometry import Point, MultiPoint
lat_col, lon_col = ('Latitude', 'Longitude')

crs='EPSG:4326' #standard GPS coordinate system

#Convert Dataframe to GeoDataFrame
df_filled['geometry'] = df_filled.apply(lambda row: Point(row[lon_col], row[lat_col]), axis=1)
gdf = gpd.GeoDataFrame(df_filled, geometry='geometry', crs=crs)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import MultiPoint
from sklearn.cluster import DBSCAN

#Extract Latitude & Longitude from Geometry
gdf['Latitude'] = gdf['geometry'].y
gdf['Longitude'] = gdf['geometry'].x

#Convert to Radians for Haversine Metric
coords = np.radians(gdf[['Latitude', 'Longitude']].values)

# Function to Plot K-Distance Graph
def plot_k_distance_graph(X, k):
    neigh = NearestNeighbors(n_neighbors=k, metric='haversine')
    neigh.fit(X)
    distances, _ = neigh.kneighbors(X)
    distances = np.sort(distances[:, k-1])  # Sort distances for k-th nearest neighbor
    plt.figure(figsize=(6, 5))
    plt.plot(distances)
    plt.xlabel('Points')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.title('K-distance Graph')
    plt.grid()
    plt.show()

#Plot the K-Distance Graph
plot_k_distance_graph(coords, k=5)  #k=5, 10, or 20 to see the elbow
```

From the k-graph above, the elbow seems to be around 0.001.
* This implies that optimal epsilon should be 0.001 x 6371 = 6.371
*Approximately 6

Resource: [DataCamp, 2024](https://www.datacamp.com/tutorial/dbscan-clustering-algorithm)

```python
# Apply DBSCAN Clustering
kms_per_radian = 6371.0088
epsilon = 6 / kms_per_radian  # 6 km radius (based on k-graph)

db = DBSCAN(eps=epsilon, min_samples=10, algorithm='ball_tree', metric='haversine').fit(coords)
gdf['Cluster'] = db.labels_  # Assign clusters

#Centermost Point for Each Cluster
clusters = gdf.groupby('Cluster')[['Latitude', 'Longitude']].apply(lambda group: MultiPoint(group.values).centroid)
gdf['Cluster_Centroid'] = gdf['Cluster'].map(clusters)

#Extract Centermost Coordinates
gdf['Center_Lat'] = gdf['Cluster_Centroid'].apply(lambda point: point.y)
gdf['Center_Lon'] = gdf['Cluster_Centroid'].apply(lambda point: point.x)

#Drop Extra Geometry Column
gdf = gdf.drop(columns=['Cluster_Centroid'])

# Display the Number of Clusters Found
num_clusters = gdf['Cluster'].nunique()
print(f'Number of clusters found: {num_clusters}')
```

### **Hyperparameter Tuninig & Remodelling**

Resource: [Ali, 2023](https://medium.com/@alihijazy/hyperparameters-in-regression-models-a-quick-dive-785b4a56bd2b)

Work flow:
1.  Decision Tree (Overfitting):
    * Perform tuning on max_depth, min_samples_split, min_samples_leaf
2. Random Forest (slight overfitting):
    * Perform tuning on max_depth, min_samples_split, n_estimators

#### Decision Tree parameter tuning

```python
depth_values = range(2, 21)

#Dictionary to store best scores and depth for each metric
best_scores = {}
best_depths = {}

for metric in ["RMSE", "MAE", "R2"]:
    best_scores[metric] = float("inf") if metric != "R2" else -float("inf")
    best_depths[metric] = None

print("Tuning Decision Tree max_depth....\n")

for depth in depth_values:
    model = DecisionTreeRegressor(max_depth=depth, random_state=42)
    model.fit(X_train_processed, y_train)
    y_pred_train = model.predict(X_train_processed)
    y_pred_test = model.predict(X_test_processed)

    train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

    train_mae = mean_absolute_error(y_train, y_pred_train)
    test_mae = mean_absolute_error(y_test, y_pred_test)

    train_r2 = r2_score(y_train, y_pred_train)
    test_r2 = r2_score(y_test, y_pred_test)

    # Errors
    error_rmse = 1 - test_rmse
    error_mae = 1 - test_mae
    error_r2 = 1 - test_r2

    print(f"Depth {depth} has the following scores:\n Test RMSE: {test_rmse:.4f}, Test MAE: {test_mae:.4f}, Test R2: {test_r2:.4f}")

    # Update best depth based on lowest error
    if test_rmse < best_scores["RMSE"]:
        best_scores["RMSE"] = test_rmse
        best_depths["RMSE"] = depth

    if test_mae < best_scores["MAE"]:
        best_scores["MAE"] = test_mae
        best_depths["MAE"] = depth

    if test_r2 > best_scores["R2"]:
        best_scores["R2"] = test_r2
        best_depths["R2"] = depth

# Print optimal depths
print("\n Best max_depth values found:")
for metric, depth in best_depths.items():
    print(f"{metric}: Best depth = {depth}")
```

from the above, the most optimal value for max_depth is 5

```python
split_values = [2, 5, 10, 20, 50]
leaf_values = [1, 2, 5, 10, 20]

# Initialize scores and corresponding parameter
best_params = {
    "RMSE": {"min_samples_split": None, "min_samples_leaf": None, "score": float("inf")},
    "MAE": {"min_samples_split": None, "min_samples_leaf": None, "score": float("inf")},
    "R2": {"min_samples_split": None, "min_samples_leaf": None, "score": -float("inf")}
}

#
for split in split_values:
    for leaf in leaf_values:
        model = DecisionTreeRegressor(min_samples_split=split, min_samples_leaf=leaf, max_depth=5)
        model.fit(X_train_processed, y_train)
        y_pred = model.predict(X_test_processed)

        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Track best scores
        if rmse < best_params["RMSE"]["score"]:
            best_params["RMSE"] = {"min_samples_split": split, "min_samples_leaf": leaf, "score": rmse}

        if mae < best_params["MAE"]["score"]:
            best_params["MAE"] = {"min_samples_split": split, "min_samples_leaf": leaf, "score": mae}

        if r2 > best_params["R2"]["score"]:
            best_params["R2"] = {"min_samples_split": split, "min_samples_leaf": leaf, "score": r2}

        print(f"Split {split}, Leaf {leaf} -> RMSE: {rmse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}")

print("\nBest min_samples_split and min_samples_leaf values found:")
for metric, params in best_params.items():
    print(f"{metric}: Best split = {params['min_samples_split']}, Best leaf = {params['min_samples_leaf']}, Score = {params['score']:.4f}")
```

From the results above, the most optimal values are split=50, leaf=20

##### Final Finding: For Decision Tree, paramters to be used include:
*     max_depth = 5
*     min_samples_split = 50
*     min_samples_leaf  = 5


### Random Forest Tuning
